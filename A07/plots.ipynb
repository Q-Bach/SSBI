```python
import matplotlib.pyplot as plt

# Input values
peaks = {
    'peak1': {'intensity': 0.55, 'mass': 1148.5153},
    'peak2': {'intensity': 0.30, 'mass': 1149.5153},
    'peak3': {'intensity': 0.10, 'mass': 1150.5153}
}

# Extract intensity and mass values
intensities = [peak['intensity'] for peak in peaks.values()]
masses = [peak['mass'] for peak in peaks.values()]

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(masses, intensities, width=0.1, align='center', color='b')
plt.xlabel('m/z')
plt.ylabel('Intensity')
plt.title('MS Peaks')
plt.xticks(masses)
plt.ylim(0, max(intensities) + 0.1)
plt.grid(True, linestyle='--')
plt.show()
```
